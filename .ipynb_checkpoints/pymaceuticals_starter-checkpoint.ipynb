```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_file = "data/mouse_drug_data.csv"
clinical_trial_data_file = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_file)
trial_df = pd.read_csv(clinical_trial_data_file)


# Combine the data into a single dataset
mouse_trial_df = pd.merge(trial_df, mouse_df, on="Mouse ID", how = 'outer')


# Display the data table for preview
mouse_trial_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mouse_trial_summary_df = mouse_trial_df[["Drug", "Timepoint", "Tumor Volume (mm3)"]]


# Convert to DataFrame
drug_group_df = pd.DataFrame(mouse_trial_summary_df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean())


# Preview DataFrame
drug_group_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint


# Convert to DataFrame
drug_group_sem = pd.DataFrame(mouse_trial_summary_df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem())


# Preview DataFrame
drug_group_sem.head()

```

```python
 # Minor Data Munging to Re-Format the Data Frames
drug_group_pivot_df = drug_group_df.pivot_table(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')      


# Preview that Reformatting worked
drug_group_pivot_df

```

```python
drug_group_sem_pivot_df = drug_group_sem.pivot_table(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')

drug_group_sem_pivot_df.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis = np.arange(0, 50, 5)

plt.errorbar(drug_group_pivot_df.index, drug_group_pivot_df['Capomulin'],  
            marker='s', yerr = drug_group_sem_pivot_df['Capomulin'])

plt.errorbar(drug_group_pivot_df.index, drug_group_pivot_df['Infubinol'],
            marker='o', yerr = drug_group_sem_pivot_df['Infubinol'])

plt.errorbar(drug_group_pivot_df.index, drug_group_pivot_df['Ketapril'], 
            marker='v', yerr = drug_group_sem_pivot_df['Ketapril'])

plt.errorbar(drug_group_pivot_df.index, drug_group_pivot_df['Placebo'], 
            marker='x', yerr = drug_group_sem_pivot_df['Placebo'])

x_lim = (-5, 50)

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc='best')

plt.grid()

plt.savefig('volume.png')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mouse_sites_summary_df = mouse_trial_df[["Drug", "Timepoint", "Metastatic Sites"]]

# Convert to DataFrame
drug_sites_df = pd.DataFrame(mouse_sites_summary_df.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean())

# Preview DataFrame
drug_sites_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
drug_sites_sem_df_pivot = drug_sites_sem_df.pivot_table(index='Timepoint',columns='Drug',values='Metastatic Sites')

# Preview that Reformatting worked
drug_sites_sem_df_pivot.head()
```

```python
# Generate the Plot (with Error Bars)



x_axis = np.arange(0, 50, 5)

plt.errorbar(drug_sites_df_pivot.index, drug_sites_df_pivot['Capomulin'],  
            marker='s', yerr = drug_sites_sem_df_pivot['Capomulin'])

plt.errorbar(drug_sites_df_pivot.index, drug_sites_df_pivot['Infubinol'],
            marker='o', yerr = drug_sites_sem_df_pivot['Infubinol'])

plt.errorbar(drug_sites_df_pivot.index, drug_sites_df_pivot['Ketapril'], 
            marker='v', yerr = drug_sites_sem_df_pivot['Ketapril'])

plt.errorbar(drug_sites_df_pivot.index, drug_sites_df_pivot['Placebo'], 
            marker='x', yerr = drug_sites_sem_df_pivot['Placebo'])

x_lim = (-5, 50)

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc='best')

plt.grid()



# Save the Figure
plt.savefig('metastatic.png')

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_df = mouse_trial_df[["Drug", "Timepoint", "Mouse ID"]]

# Convert to DataFrame
mouse_count_df = pd.DataFrame(mouse_df.groupby(['Drug', 'Timepoint'])['Mouse ID'].count())

# Preview DataFrame
mouse_count_df.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_pivot = mouse_count_df.pivot_table(index='Timepoint',columns='Drug',values='Mouse ID')

# Preview the Data Frame
mouse_count_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)


x_axis = np.arange(0, 50, 5)


y1 = mouse_count_pivot["Capomulin"] / mouse_count_pivot["Capomulin"][0] * 100
y2 = mouse_count_pivot["Infubinol"] / mouse_count_pivot["Infubinol"][0] * 100
y3 = mouse_count_pivot["Ketapril"] / mouse_count_pivot["Ketapril"][0] * 100
y4 = mouse_count_pivot["Placebo"] / mouse_count_pivot["Placebo"][0] * 100


plt.plot(mouse_count_pivot.index, y1,  
            marker='s')

plt.plot(mouse_count_pivot.index, y2,  
            marker='o')

plt.plot(mouse_count_pivot.index, y3,  
            marker='v')

plt.plot(mouse_count_pivot.index, y4,  
            marker='x')


plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend(loc='best')

plt.grid()


# Save the Figure
plt.savefig('survival.png')

# Show the Figure
plt.show()
```
